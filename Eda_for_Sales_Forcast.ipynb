# Hypothesis Testing


```python
from scipy.stats import ttest_ind, f_oneway
from scipy import stats
```

```python
sales_L3 = data[data['Location_Type'] == 'L3']['Sales'].values
sales_L2 = data[data['Location_Type'] == 'L2']['Sales'].values
sales_L1 = data[data['Location_Type'] == 'L1']['Sales'].values
sales_L5 = data[data['Location_Type'] == 'L5']['Sales'].values
sales_L4 = data[data['Location_Type'] == 'L4']['Sales'].values
```

```python
f_stat, p_value = stats.f_oneway(sales_L3, sales_L2, sales_L1, sales_L5, sales_L4)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
sales_S1 = data[data['Store_Type'] == 'S1']['Sales'].values
sales_S2 = data[data['Store_Type'] == 'S2']['Sales'].values
sales_S3 = data[data['Store_Type'] == 'S3']['Sales'].values
sales_S4 = data[data['Store_Type'] == 'S4']['Sales'].values
```

```python
f_stat, p_value = stats.f_oneway(sales_S1, sales_S2, sales_S3, sales_S4)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
holiday_sales = data[data['Holiday'] == 1]['Sales'].values
workingday_sales = data[data['Holiday'] == 0]['Sales'].values
```

```python
f_stat, p_value = stats.f_oneway(holiday_sales, workingday_sales)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
Discount_sales = data[data['Discount'] == 'Yes']['Sales'].values
NOnDiscount_sales = data[data['Discount'] == 'No']['Sales'].values
```

```python
f_stat, p_value = stats.f_oneway(Discount_sales, NOnDiscount_sales)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

# Feature Engineering


```python
import pandas as pd

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['DayOfWeek'] = data['Date'].dt.dayofweek
data['IsWeekend'] = data['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)

# Holiday and Discount as binary features are already in the dataset
# Creating lag features could be useful if you are working with time series models
```

```python
# Group by 'Date' and aggregate sales
daily_sales = data.groupby('Date').agg({
    'Sales': 'sum'  # Sum up sales for each date
}).reset_index()

# Set Date as the index for time series analysis
daily_sales['Date'] = pd.to_datetime(daily_sales['Date'])
daily_sales.set_index('Date', inplace=True)
```

# Model Building

```python
min_date = daily_sales.index.min()
max_date = daily_sales.index.max()

print(f"Date range in the dataset: {min_date} to {max_date}")
```

```python
# Let's use the first 15 months for training and the last 2 months for testing
split_date = '2019-04-01'

# Split the data
train_data = daily_sales[daily_sales.index < split_date]
test_data = daily_sales[daily_sales.index >= split_date]
```

```python
train_data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(train_data['Sales'])
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(train_data['Sales'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
model = ARIMA(train_data['Sales'], order=(1, 0, 0))
model_fit = model.fit()
forecast_steps = len(test_data)
forecast = model_fit.forecast(steps=forecast_steps)

# Directly convert to DataFrame if it's a Series
forecast_df = forecast.to_frame(name='Forecast')

# Check the contents of forecast_df
forecast_df

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate the evaluation metrics
mae = mean_absolute_error(test_data['Sales'], forecast_df['Forecast'])
mse = mean_squared_error(test_data['Sales'], forecast_df['Forecast'])
rmse = mean_squared_error(test_data['Sales'], forecast_df['Forecast'], squared=False)  # squared=False for RMSE
mape = np.mean(np.abs((test_data['Sales'] - forecast_df['Forecast']) / test_data['Sales'])) * 100  # MAPE

# Print the results
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```

```python
# Assuming 'test_data' has the actual sales and 'forecast_df' has the forecasted sales
comparison_df = pd.DataFrame({
    'Actual': test_data['Sales'],   # Replace 'Sales' with the actual sales column
    'Forecast': forecast_df['Forecast']
})

# Check the head of the comparison DataFrame
print(comparison_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for better visuals
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))

# Plot actual sales
plt.plot(comparison_df.index, comparison_df['Actual'], label='Actual Sales', color='blue')

# Plot forecasted sales
plt.plot(comparison_df.index, comparison_df['Forecast'], label='Forecasted Sales', color='orange', linestyle='--')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Actual vs Forecasted Sales')
plt.legend()

# Show the plot
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pandas as pd

# Assuming you have train_data and test_data as your datasets
# Define the seasonal order; adjust these values based on your data
seasonal_order = (1, 1, 2, 30)  # Example seasonal order for daily data with a 30-day cycle

# Fit the SARIMA model
sari_model = SARIMAX(train_data['Sales'], order=(1, 0, 1), seasonal_order=seasonal_order)
sari_model_fit = sari_model.fit()

# Forecasting
sari_forecast_steps = len(test_data)
sari_forecast = sari_model_fit.forecast(steps=sari_forecast_steps)

# Convert to DataFrame if it's a Series
sari_forecast_df = sari_forecast.to_frame(name='Forecast')

# Check the contents of forecast_df
print(sari_forecast_df)
```

```python
# Assuming 'test_data' has the actual sales and 'forecast_df' has the forecasted sales
sari_comparison_df = pd.DataFrame({
    'Actual': test_data['Sales'],   # Replace 'Sales' with the actual sales column
    'Forecast': sari_forecast_df['Forecast']
})

# Check the head of the comparison DataFrame
print(sari_comparison_df.head())
```
